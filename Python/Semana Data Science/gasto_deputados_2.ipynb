## Quantos deputados/ agentes públicos tem por partido?

```python
partidos = data_2018.groupby(['sgPartido'])['txNomeParlamentar'].nunique()
```

## Gasto por partido por agentes públicos

```python
gasto = data_2018.groupby(['sgPartido'])['vlrDocumento'].sum()
```

```python
df = pd.concat([gasto, partidos], axis=1)
df["gastosDeputados"] = df["vlrDocumento"] / df["txNomeParlamentar"]
df.sort_values(ascending=False, by='gastosDeputados')[:10]
```

```python
deputados = pd.read_csv("data/deputados.csv", 
                        low_memory=False, sep=';', 
                        parse_dates=["dataNascimento", "dataFalecimento"])
```

```python
deputados = deputados[(deputados['idLegislaturaInicial'] == 56) | (deputados['idLegislaturaFinal'] == 56)]
```

```python
deputados2 = deputados.drop(['idLegislaturaInicial', 
                             'cpf', 'urlRedeSocial', 
                             'dataFalecimento', 
                             'urlWebsite', 
                             'urlRedeSocial',
                             'idLegislaturaFinal', 
                             'uri'], axis=1)
deputados2
```

```python
deputados_iguais = []
for linha_dado_2018 in data_2019['txNomeParlamentar'].unique():
    for linha_deputado2 in deputados2['nome'].unique():
        if(linha_dado_2018 == linha_deputado2):
            deputados_iguais.append(linha_dado_2018)
len(deputados_iguais)
```

```python
from difflib import SequenceMatcher
```

```python
deputados_iguais = []
for linha_dado_2018 in data_2019['txNomeParlamentar'].unique():
    for linha_deputado2 in deputados2['nome'].unique():
        s = SequenceMatcher(None, linha_dado_2018, linha_deputado2)
        if(s.ratio() > 0.86 ):
            deputados_iguais.append(linha_dado_2018)
len(deputados_iguais)
```
