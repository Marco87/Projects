## Deputados

```python
from difflib import SequenceMatcher
```

```python
deputados = pd.read_csv("Dados/deputados.csv", sep=";", parse_dates=['dataNascimento', 'dataFalecimento'])
```

```python
deputados = deputados[(deputados['idLegislaturaInicial'] == 56) | (deputados['idLegislaturaFinal'] == 56)]
```

```python
deputados2 = deputados.drop(['idLegislaturaInicial', 'cpf', 'urlRedeSocial', 'dataFalecimento', 'urlWebsite', 'urlRedeSocial',
                     'idLegislaturaFinal', 'uri'], axis=1)
```

```python
deputados_iguais = []
for linha_dado_2018 in data_2018['txNomeParlamentar'].unique():
    for linha_deputado2 in deputados2['nome'].unique():
        if(linha_dado_2018 == linha_deputado2):
            #print(f"{linha_dado_2018} == {linha_deputado2}")
            deputados_iguais.append(linha_dado_2018)
len(deputados_iguais)
```

```python
deputados_iguais = []
for linha_dado_2018 in data_2019['txNomeParlamentar'].unique():
    for linha_deputado2 in deputados2['nome'].unique():
        s = SequenceMatcher(None, linha_dado_2018, linha_deputado2)
        if(s.ratio() > 0.86 ):
            #print(f"{linha_dado_2018} == {linha_deputado2}")
            deputados_iguais.append(linha_dado_2018)
len(deputados_iguais)
```

```python
gastos_genero = []
for nomes_2019 in data_2019:
    for dep in deputados:
        if nomes_2019.loc['txNomeParlamentar'] == dep.loc['nome']:
            gastos_genero.append(nomes_2019)
print(len(gastos_genero))
```
