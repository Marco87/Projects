```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 7, 1, 0]
titulo = "Gráfico de barras"
eixox = "Eixo X"
eixoy = "Eixo Y"

#Legendas
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.plot(x, y)
plt.show()
```

```python
plt.bar(x, y)
```

```python
x1 = [1, 3, 5, 7, 9]
y1 = [2, 3, 7, 1, 0]

x2 = [2, 4, 6, 8, 10]
y2 = [5, 1, 3, 7, 4]

#Título
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.bar(x1, y1, label="Grupo 1", color="red")
plt.bar(x2, y2, label="Grupo 2", color="black")
plt.legend()
plt.show()
```

```python
#Título
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.plot(x1, y1, label="Grupo 1", color="red", linestyle='-.')
plt.scatter(x1, y1, label="Grupo 2", color="black", s=500)
plt.legend()
plt.show()
#plt.savefig("figura1.png") -> salva a imagem gerada
```
