```python
import matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Importando a base de dados para o nosso projeto

Os dados abaixo de referem às informações 

```python
#Base de dados completa, com os gols de Messi e Ronaldo entre as temporadas 2009-2010 e 2017-2018

df_messironaldo = pd.read_excel('data/MessiRonaldo.xlsx')
```

```python
#Base de dados com informações de Messi e Cristiano Ronaldo

df_messironaldo
```

```python
#Dados do Lionel Messi

df_messi = df_messironaldo.query('Player == "Messi"')
```

```python
df_messi
```

```python
#Dados do Cristiano Ronaldo

df_ronaldo = df_messironaldo.query('Player == "Ronaldo"')
```

```python
df_ronaldo
```

### Champions League

Comparação entre os dois levando apenas em consideração os gols na Champions League

![](https://logospng.org/download/uefa-champions-league/logo-uefa-champions-league-256.png)

```python
#Removendo as features referentes a La Liga (Lionel Messi)

df_messi_champions = df_messi.drop(columns=['Liga_Goals', 'Liga_Asts', 'Liga_Aps', 'Liga_Mins', 'Liga_Team_Goals'])
```

```python
#Removendo as features referentes a La Liga (Cristiano Ronaldo)

df_ronaldo_champions = df_ronaldo.drop(columns=['Liga_Goals', 'Liga_Asts', 'Liga_Aps', 'Liga_Mins', 'Liga_Team_Goals'])
```

```python
#Gols do Lionel Messi na Champions League

plt.barh(df_messi_champions['Season'], df_messi_champions['CL_Goals'], color="red")

plt.title("Gols do Lionel Messi na Champions League")
```

```python
#Gols do Cristiano Ronaldo na Champions League

plt.barh(df_ronaldo_champions['Season'], df_ronaldo['CL_Goals'], color='blue')

plt.title("Gols do Cristiano Ronaldo na Champions League")
```

```python
#Mesclando as bases de Messi e Ronaldo (dados da Champions League)

df_messi_ronaldo_champions = df_messi_champions.merge(df_ronaldo_champions, on=('Season'), suffixes=('_Messi', '_Ronaldo'))
```

```python
#Ordenando os dados pela coluna 'Season' (do menor ano para o mauir)

df_messi_ronaldo_champions = df_messi_ronaldo_champions.sort_values(by=['Season'], ascending=True)
```
