### Importando as bibliotecas necessárias

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Importando a base de dados para o nosso projeto

Os dados abaixo de referem às informações 

```python
#Base de dados completa, com os gols de Messi e Ronaldo entre as temporadas 2009-2010 e 2017-2018

df_messironaldo = pd.read_excel('data/MessiRonaldo.xlsx')
```

```python
#Base de dados com informações de Messi e Cristiano Ronaldo

df_messironaldo
```

```python
#Dados do Lionel Messi

df_messi = df_messironaldo.query('Player == "Messi"')
```

```python
df_messi
```

```python
#Dados do Cristiano Ronaldo

df_ronaldo = df_messironaldo.query('Player == "Ronaldo"')
```

```python
df_ronaldo
```

### La Liga

Comparação entre os dois levando apenas em consideração os gols em La Liga

![](https://cdn.clipart.email/aa04921fb81612f5c87869e0872191da_la-liga-pes-2020-leagues-competitions-pro-evolution-soccer-_256-256.png)

```python
#Removendo as features referentes a Liga dos Campeões (Lionel Messi)

df_messi_liga = df_messi.drop(columns=['CL_Goals', 'CL_Asts', 'CL_Aps', 'CL_Mins', 'CL_Team_Goals'])
```

```python
#Renomeando as features

#df_messi_liga = df_messi_liga.rename(columns={'Liga_Goals':'Liga_Goals_Messi', 'Liga_Asts':'Liga_Asts_Messi', 'Liga_Aps':'Liga_Aps_Messi', 'Liga_Mins':'Liga_Mins_Messi', 'Liga_Team_Goals':'Liga_Team_Goals_Messi'})
```

```python
#Removendo as features referentes a Liga dos Campeões (Cristiano Ronaldo)

df_ronaldo_liga = df_ronaldo.drop(columns=['CL_Goals', 'CL_Asts', 'CL_Aps', 'CL_Mins', 'CL_Team_Goals'])
```

```python
#Renomeando as features

#df_ronaldo_liga = df_ronaldo_liga.rename(columns={'Liga_Goals':'Liga_Goals_Ronaldo', 'Liga_Asts':'Liga_Asts_Ronaldo', 'Liga_Aps':'Liga_Aps_Ronaldo', 'Liga_Mins':'Liga_Mins_Ronaldo', 'Liga_Team_Goals':'Liga_Team_Goals_Ronaldo'})
```

```python
plt.bar(df_messi_liga['Season'].sort_values(ascending=True), df_messi_liga['Liga_Goals_Messi'], color='red')

#Título do gráfico
plt.title("Gols do Lionel Messi na La Liga")

#Título do eixo Y
plt.ylabel('Gols')

#Rotação dos elementos do eixo X
plt.xticks(rotation=45)

plt.show()
```

```python
plt.bar(df_ronaldo_liga['Season'].sort_values(ascending=True), df_ronaldo_liga['Liga_Goals_Ronaldo'], color='blue')

plt.title("Gols do Cristiano Ronaldo na La Liga")
plt.ylabel('Gols')
plt.xticks(rotation=45)

plt.show()
```

```python
#Mesclando as bases de Messi e Ronaldo (dados da La Liga)

df_messi_ronaldo_liga = df_messi_liga.merge(df_ronaldo_liga, on=('Season'), suffixes=('_Messi', '_Ronaldo'))
```

```python
#Ordenando os dados pela coluna 'Season' (do menor ano para o mauir)

df_messi_ronaldo_liga = df_messi_ronaldo_liga.sort_values(by=['Season'], ascending=True)
```

```python
df_messi_ronaldo_liga.plot.bar(x='Season', y=['Liga_Goals_Messi', 'Liga_Goals_Ronaldo'], rot=45, width=.7)
```
