### Importando as bibliotecas necessárias

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Importando a base de dados para o nosso projeto

Os dados abaixo de referem às informações 

```python
#Base de dados completa, com os gols de Messi e Ronaldo entre as temporadas 2009-2010 e 2017-2018

df_messironaldo = pd.read_excel('data/MessiRonaldo.xlsx')
```

```python
#Base de dados com informações de Messi e Cristiano Ronaldo

df_messironaldo
```

```python
#Dados do Lionel Messi

df_messi = df_messironaldo.query('Player == "Messi"')
```

```python
df_messi
```

```python
#Dados do Cristiano Ronaldo

df_ronaldo = df_messironaldo.query('Player == "Ronaldo"')
```

```python
df_ronaldo
```

### La Liga

Comparação entre os dois levando apenas em consideração os gols em La Liga

![](https://cdn.clipart.email/aa04921fb81612f5c87869e0872191da_la-liga-pes-2020-leagues-competitions-pro-evolution-soccer-_256-256.png)

```python
#Removendo as features referentes a Liga dos Campeões (Lionel Messi)

df_messi_liga = df_messi.drop(columns=['CL_Goals', 'CL_Asts', 'CL_Aps', 'CL_Mins', 'CL_Team_Goals'])
```

```python
#Removendo as features referentes a Liga dos Campeões (Cristiano Ronaldo)

df_ronaldo_liga = df_ronaldo.drop(columns=['CL_Goals', 'CL_Asts', 'CL_Aps', 'CL_Mins', 'CL_Team_Goals'])
```

```python
plt.bar(df_messi_liga['Season'].sort_values(ascending=True), df_messi_liga['Liga_Goals'], color='red')

#Título do gráfico
plt.title("Gols do Lionel Messi na La Liga")

#Título do eixo Y
plt.ylabel('Gols')

#Rotação dos elementos do eixo X
plt.xticks(rotation=45)

plt.show()
```

```python
plt.bar(df_ronaldo_liga['Season'].sort_values(ascending=True), df_ronaldo_liga['Liga_Goals'])

plt.title("Gols do Cristiano Ronaldo na La Liga")
plt.ylabel('Gols')
plt.xticks(rotation=45)

plt.show()
```

```python
#Expessura das colunas
width=0.4

bar_messi = plt.bar(df_messi_liga['Season'].sort_values(ascending=True), df_messi_liga['Liga_Goals'], width, color='red', label="Messi")
bar_ronaldo = plt.bar(df_ronaldo_liga['Season'].sort_values(ascending=True), df_ronaldo_liga['Liga_Goals'], width, color='blue', label="Ronaldo")

#Título do gráfico
plt.title("Gols - Messi vs Ronaldo - La Liga")

#Título do eixo Y
plt.ylabel('Gols')

#Rotação dos elementos do eixo X
plt.xticks(rotation=45)

#Exibe as labels no gráfico
plt.legend()

#Estilo do gráfico
plt.style.use('ggplot')

plt.show()
```
