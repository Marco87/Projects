```python
import numpy as np
import pandas as pd
import requests
import json
```

```python
url = "https://brasil.io/api/dataset/covid19/caso/data?format=json&page=2"
```

```python
JSONContent = requests.get(url).json()
```

```python
content = json.dumps(JSONContent, indent=4, sort_keys=True)
```

```python
print(content)
```

```python
data = json.loads(content)
```

```python
df = pd.DataFrame(data)
```
